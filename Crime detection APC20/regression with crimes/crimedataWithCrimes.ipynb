Пропущенных данных довольно много (1872 из 2215 строк = 84.5% данных) , для их восполнения мы можем использовать mean imputing. Проверим, по каким колоннам мы могли бы группировать данные

```python
print("Community Name missing values: ", df_crime['communityName'].isna().sum())
print("State missing values: ", df_crime['state'].isna().sum())
print("countyCode missing values: ", df_crime['countyCode'].isna().sum())
print("communityCode missing values: ", df_crime['communityCode'].isna().sum())
```

Поскольку county code и community code имеют пропущенные значения, лучше группировать по штатам и высчитывать среднее значение по штату для imputing-а. 

```python
df_crime.iloc[:, 4:] = df_crime.iloc[:, 4:].apply(pd.to_numeric, errors='coerce')
```

```python
transform_numeric_cols = df_crime.iloc[:, 4:].columns

df_crime[transform_numeric_cols] = df_crime.groupby('state')[transform_numeric_cols].transform(lambda x: x.fillna(x.mean()))
```

Теперь восполним пропущенные значения в countyCode и communityCode

```python
len(df_crime['countyCode'].unique())
```

"Округ" имеет 625 пропущенных значений, однако мы не можем просто удалить всю колонну или строки. Удаление колонны может привести к потере важной информации, поскольку оно имеет 121 уникальное значение, это говорит нам о том,что данные действительно группируются по округам (в среднем 18 сообществ на округ). 625 строк из 2215 = 28% всех данных, поэтому мы не можем удалить строки. <br>
Наше решение - восполнить значение "округ" случайным значением из того же штата, таким образом географически эти сообщества расположены не настолько далеко, чтобы значительно изменить данные. Если по штату нет округов вообще, присваиваем "округу" значение 0. 

```python
import numpy as np

def fill_county_code(row, df):
    if pd.isna(row['countyCode']):
        non_nan_codes = df[df['state'] == row['state']]['countyCode'].dropna().values
        if len(non_nan_codes) > 0:
            return np.random.choice(non_nan_codes)
        else:
            return 0
    else:
        return row['countyCode']

    
df_crime['countyCode'] = df_crime.apply(lambda row: fill_county_code(row, df_crime), axis=1)
```

Перейдём к communityCode

```python
len(df_crime['communityCode'].unique())
```

Как мы видим, код сообществ - это не id для каждого сообщества - иначе их было бы 2215. Тем не менее, классов слишком много, и они не могут представлять большую предсказательную способность для данных, в которых 2000 строк. Удалим эту колонну.

```python
df_crime = df_crime.drop(['communityCode'], axis=1)
```

Восполним оставшиеся значения средним значением по колонне, поскольку удаление целых штатов было бы опрометчиво, как и удаление нескольких колонн, которые могут быть таргетом. 

```python
df_crime = df_crime.apply(lambda col: col.fillna(col.mean()) if col.isna().any() else col)
```

## Получив чистый датасет, мы можем создать функцию, которая будет универсальной для ВСЕХ таргетов. Алгоритм: выбор таргета, удаление аутлаеров в данном таргете, применение target encoding-а для классовых колонн (штат, округ), деление на тренировочные и тестовые данные, feature selection / выбор приоритетных колонн, тренировка моделей, сохранение лучшей модели.

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from xgboost import XGBRegressor
from sklearn.metrics import root_mean_squared_error, mean_absolute_error, r2_score
import joblib
import os
import json
import matplotlib.pyplot as plt

os.makedirs("imagesWithCrimes", exist_ok=True)
os.makedirs("feature_weightsWithCrimes", exist_ok=True)
os.makedirs("modelsWithCrimes", exist_ok=True)
os.makedirs("encoded_mappingsWithCrimes", exist_ok=True)

def train_models_for_targets(data, targets):
    results = {}

    
    for target in targets:
        print(f"Processing target: {target}")

        X = data.drop(columns= [target,
    "murdPerPop",
    "rapesPerPop",
    "robbbPerPop",
    "assaultPerPop",
    "burglPerPop",
    "larcPerPop",
    "autoTheftPerPop",
    "arsonsPerPop",
    "ViolentCrimesPerPop",
    "nonViolPerPop", 'communityName'], errors='ignore')
        y = data[target]

        # Удаление аутлауеров / выбросов
        Q1 = y.quantile(0.25)
        Q3 = y.quantile(0.75)
        IQR = Q3 - Q1
        data_filtered = data[(y >= Q1 - 1.5 * IQR) & (y <= Q3 + 1.5 * IQR)]
        y_filtered = data_filtered[target]
        X = X.loc[data_filtered.index]

        # Таргет энкодинг / преобразование классов в численные значения для модели
        categorical_columns = ["state", "countyCode"]
        
        # вычисление средних значений по штатам и округам для преобразования
        state_target_encoding = data_filtered.groupby('state')[target].mean()
        county_target_encoding = data_filtered.groupby('countyCode')[target].mean()

        # Добавляем полученные значения в качестве новых колонн в новый датафрейм
        data_filtered['encoded_state'] = data_filtered['state'].map(state_target_encoding)
        data_filtered['encoded_county'] = data_filtered['countyCode'].map(county_target_encoding)

        # Сохраняем полученные преобразованные данные для работы с бэкендом в дальнейшем
        state_mapping = pd.DataFrame({
            'State': state_target_encoding.index,
            'Mean_Target': state_target_encoding.values
        })
        county_mapping = pd.DataFrame({
            'CountyCode': county_target_encoding.index,
            'Mean_Target': county_target_encoding.values
        })

        state_mapping.to_csv(f"encoded_mappingsWithCrimes/{target}_state_mapping.csv", index=False)
        county_mapping.to_csv(f"encoded_mappingsWithCrimes/{target}_county_mapping.csv", index=False)

        # Добавляем колонны в число тренировочных данных, удаляем старые
        X['encoded_state'] = data_filtered['encoded_state']
        X['encoded_county'] = data_filtered['encoded_county']
        X = X.drop(columns = categorical_columns)
        
        # деление на тренировочные и тестовые данные 
        X_train, X_test, y_train, y_test = train_test_split(X, y_filtered, test_size=0.2, random_state=42)

        # Выбираем приоритетные колонны, топ-20 используя случайный лес 
        selector = RandomForestRegressor(random_state=42)
        selector.fit(X_train, y_train)
        importances = selector.feature_importances_
        top_features_idx = np.argsort(importances)[-20:]
        top_features = X.columns[top_features_idx]

        X_train = X_train[top_features]
        X_test = X_test[top_features]

        # Сохраним информацию о приоритетных колоннах 
        feature_weights = pd.DataFrame({"Feature": top_features, "Importance": importances[top_features_idx]})
        feature_weights.sort_values(by="Importance", ascending=False, inplace=True)
        feature_weights.to_csv(f"feature_weightsWithCrimes/{target}.csv", index=False)

        # Сохраним также в качестве графика для визуализации на сайте 
        plt.figure(figsize=(10, 6))
        plt.barh(feature_weights["Feature"], feature_weights["Importance"], color="skyblue")
        plt.xlabel("Importance")
        plt.ylabel("Feature")
        plt.title(f"Feature Importance for {target}")
        plt.savefig(f"imagesWithCrimes/{target}.jpg")
        plt.close()
        print(X_train.columns)

        # Тренировка моделей 
        models = {
            "RandomForest": RandomForestRegressor(random_state=42),
            "KNN": KNeighborsRegressor(),
            "XGBoost": XGBRegressor(random_state=42),
            "DecisionTree": DecisionTreeRegressor(random_state=42)
        }

        metrics = {}
        for model_name, model in models.items():
            model.fit(X_train, y_train)
            y_pred = model.predict(X_test)

            # Вычисление метрик 
            rmse = root_mean_squared_error(y_test, y_pred)
            mae = mean_absolute_error(y_test, y_pred)
            r2 = r2_score(y_test, y_pred)

            metrics[model_name] = {"RMSE": rmse, "R2": r2, "MAE": mae}

        # Выбираем лучшую модель основываясь на самом низком RMSE
        best_model_name = min(metrics, key=lambda m: metrics[m]["RMSE"])
        best_model = models[best_model_name]
        print(f"Best model for {target}: {best_model_name} with RMSE {metrics[best_model_name]['RMSE']}")

        # Сохраняем модель в расширении pkl чтобы подключить к бэкенду 
        joblib.dump(best_model, f"modelsWithCrimes/{target}.pkl")

        # Добавляем полученные результаты в json файл
        results[target] = {
            "BestModel": best_model_name,
            "Mean": y.mean(),  
            **metrics[best_model_name]
        }

    # Сохраняем 
    results_path = "resultsWithCrimes.json"
    with open(results_path, "w") as f:
        json.dump(results, f, indent=4)
    print(f"Results saved to {results_path}")
```

Определяем таргеты и запускаем функцию 

```python
targets = [
    "murders",
    "rapes",
    "robberies",
    "assaults",
    "burglaries",
    "larcenies",
    "autoTheft",
    "arsons"
]
train_models_for_targets(df_crime, targets)
```
